```python
from bs4 import BeautifulSoup

# Assuming `soup` is your parsed HTML
tbody = soup.find('tbody')  # Find the <tbody> tag
problem_links = tbody.find_all('a', href=lambda x: x and x.startswith('/problem/'))

for link in problem_links:
    problem_name = link.get_text()  # Extract the problem name
    problem_url = link.get('href')  # Extract the problem link
    problem_number = problem_url.split('/')[2]  # Extract the problem number
    print(f"//백준#{problem_number} {problem_name} https://www.acmicpc.net{problem_url}")

from datetime import datetime

# Get the current time
current_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

print(current_time)
```
